```python
import pandas as pd
import os
```

```python
os.listdir('./BCCA_aperio_results/')
```

```python
tfl_a = pd.read_csv('./BCCA_aperio_results/FOXP3 Data on tFL Block A.csv')
xtfl_a = pd.read_csv('./BCCA_aperio_results/FOXP3 Data on xtrFL Block A.csv')
tfl_b = pd.read_csv('./BCCA_aperio_results/FOXP3 Data on tFL Block B.csv')
xtfl_b = pd.read_csv('./BCCA_aperio_results/FOXP3 Data on xtrFL Block B.csv')
tfl_c = pd.read_csv('./BCCA_aperio_results/FOXP3 Data on tFL Block C.csv')
xtfl_c = pd.read_csv('./BCCA_aperio_results/FOXP3 Data on xtrFL Block C.csv')
tfl_d = pd.read_csv('./BCCA_aperio_results/FOXP3 Data on tFL Block D.csv')
tfl_e = pd.read_csv('./BCCA_aperio_results/FOXP3 Data on tFL Block E.csv')
tfl_f = pd.read_csv('./BCCA_aperio_results/FOXP3 Data on tFL Block F.csv')
tfl_g = pd.read_csv('./BCCA_aperio_results/FOXP3 Data on tFL Block G.csv')
tfl_h = pd.read_csv('./BCCA_aperio_results/FOXP3 Data on tFL Block H.csv')
```

```python


#tfl_a
pos = tfl_a.columns[-2]
tfl_a = tfl_a.loc[:,['Row','Col','Sector',pos]]
tfl_a.rename(columns={pos:'Positivity'}, inplace=True)
block = ['A' for elem in range(len(tfl_a.Col.values))]
grouping = ['TFL' for elem in range(len(tfl_a.Col.values))]
tfl_a['Block'] = block
tfl_a['Grouping'] = grouping

#tfl_b
pos = tfl_b.columns[-2]
tfl_b = tfl_b.loc[:,['Row','Col','Sector',pos]]
tfl_b.rename(columns={pos:'Positivity'}, inplace=True)
block = ['B' for elem in range(len(tfl_b.Col.values))]
grouping = ['TFL' for elem in range(len(tfl_b.Col.values))]
tfl_b['Block'] = block
tfl_b['Grouping'] = grouping

#tfl_c
pos = tfl_c.columns[-2]
tfl_c = tfl_c.loc[:,['Row','Col','Sector',pos]]
tfl_c.rename(columns={pos:'Positivity'}, inplace=True)
block = ['C' for elem in range(len(tfl_c.Col.values))]
grouping = ['TFL' for elem in range(len(tfl_c.Col.values))]
tfl_c['Block'] = block
tfl_c['Grouping'] = grouping

#tfl_d
pos = tfl_d.columns[-2]
tfl_d = tfl_d.loc[:,['Row','Col','Sector',pos]]
tfl_d.rename(columns={pos:'Positivity'}, inplace=True)
block = ['D' for elem in range(len(tfl_d.Col.values))]
grouping = ['TFL' for elem in range(len(tfl_d.Col.values))]
tfl_d['Block'] = block
tfl_d['Grouping'] = grouping

#tfl_e
pos = tfl_e.columns[-2]
tfl_e = tfl_e.loc[:,['Row','Col','Sector',pos]]
tfl_e.rename(columns={pos:'Positivity'}, inplace=True)
block = ['E' for elem in range(len(tfl_e.Col.values))]
grouping = ['TFL' for elem in range(len(tfl_e.Col.values))]
tfl_e['Block'] = block
tfl_e['Grouping'] = grouping

#tfl_f
pos = tfl_f.columns[-2]
tfl_f = tfl_f.loc[:,['Row','Col','Sector',pos]]
tfl_f.rename(columns={pos:'Positivity'}, inplace=True)
block = ['F' for elem in range(len(tfl_f.Col.values))]
grouping = ['TFL' for elem in range(len(tfl_f.Col.values))]
tfl_f['Block'] = block
tfl_f['Grouping'] = grouping

#tfl_g
pos = tfl_g.columns[-2]
tfl_g = tfl_g.loc[:,['Row','Col','Sector',pos]]
tfl_g.rename(columns={pos:'Positivity'}, inplace=True)
block = ['G' for elem in range(len(tfl_g.Col.values))]
grouping = ['TFL' for elem in range(len(tfl_g.Col.values))]
tfl_g['Block'] = block
tfl_g['Grouping'] = grouping

#tfl_h
pos = tfl_h.columns[-2]
tfl_h = tfl_h.loc[:,['Row','Col','Sector',pos]]
tfl_h.rename(columns={pos:'Positivity'}, inplace=True)
block = ['H' for elem in range(len(tfl_h.Col.values))]
grouping = ['TFL' for elem in range(len(tfl_h.Col.values))]
tfl_h['Block'] = block
tfl_h['Grouping'] = grouping

#xtrfl_a
pos = xtfl_a.columns[-2]
xtfl_a = xtfl_a.loc[:,['Row','Col','Sector',pos]]
xtfl_a.rename(columns={pos:'Positivity'}, inplace=True)
block = ['A' for elem in range(len(xtfl_a.Col.values))]
grouping = ['XTRFL' for elem in range(len(xtfl_a.Col.values))]
xtfl_a['Block'] = block
xtfl_a['Grouping'] = grouping

#xtrfl_b
pos = xtfl_b.columns[-2]
xtfl_b = xtfl_b.loc[:,['Row','Col','Sector',pos]]
xtfl_b.rename(columns={pos:'Positivity'}, inplace=True)
block = ['B' for elem in range(len(xtfl_b.Col.values))]
grouping = ['XTRFL' for elem in range(len(xtfl_b.Col.values))]
xtfl_b['Block'] = block
xtfl_b['Grouping'] = grouping

#xtrfl_c
pos = xtfl_c.columns[-2]
xtfl_c = xtfl_c.loc[:,['Row','Col','Sector',pos]]
xtfl_c.rename(columns={pos:'Positivity'}, inplace=True)
block = ['C' for elem in range(len(xtfl_c.Col.values))]
grouping = ['XTRFL' for elem in range(len(xtfl_c.Col.values))]
xtfl_c['Block'] = block
xtfl_c['Grouping'] = grouping

dfs = [tfl_a, tfl_b, tfl_c, tfl_d, tfl_e, tfl_f, tfl_g, tfl_h, xtfl_a, xtfl_b, xtfl_c]
aperio_df = pd.concat(dfs)
```
